# Assignment: PCA

### Create Data

```python
#dummy regression data
import numpy as np
from sklearn.datasets import make_regression
X, y = make_regression(n_features=64, effective_rank =15,random_state=0, shuffle=False, n_samples=200)
```

### E3.1 compute PCA
see https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

```python
import numpy as np
from sklearn.decomposition import PCA
```

### E3.2 Plot Eigen Values

```python
import matplotlib.pyplot as plt

pca = PCA()
pca.fit(X)

plt.plot(pca.explained_variance_)
plt.xlabel("Eigen Value Number")
plt.ylabel("Eigen Value")
plt.show()
```

### E3.3 select a $\lambda$ cut-off and project to new base

```python
# selecting the top 10 eigenvalues
pca_new = PCA(n_components=10)
X_projected = pca_new.fit_transform(X)

print("Original data shape:", X.shape)
print("Projected data shape:", X_projected.shape)
```
